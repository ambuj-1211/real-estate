## The feature are taken as input and put into predict function in the same order as below. 

```python
X=dfs[["locality_lable","type","latitude","longitude","lease_type","gym","lift","swimming_pool","parking","property_size","bathroom","facing","water_supply","balconies","LIFT","INTERNET","AC","INTERCOM","POOL","FS","SECURITY","SC","PARK","HK","PB"]]
y=dfs["rent"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
len(y_train)
```

```python
from sklearn import linear_model
reg=linear_model.LinearRegression()
reg.fit(X_train,y_train)
```

```python
y_pred=reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
